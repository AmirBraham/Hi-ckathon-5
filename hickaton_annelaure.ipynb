```python
#data caracteristics
#create a function that determines the dataset caracteristics
def caracteristics (dataset):
    carac=pd.DataFrame(index=dataset.columns.tolist())
  #Types de variables
    carac['types']=dataset.dtypes.tolist()
  #Nombre de modalites
    modalites=[]
    for i in dataset.columns.tolist():
        modalites.append(len(set(dataset[i].tolist())))
    carac['nombre de modalites']=modalites
  #Nombre de valeurs manquantes
    missing_values=[]
    for i in dataset.columns.tolist():
        missing_values.append(dataset[i].isna().sum())
    carac['number of missing values']=missing_values
  #Pourcentage de valeurs manquantes
    prop_missing_values=[]
    for i in dataset.columns.tolist():
        prop_missing_values.append(round(((dataset[i].isna().sum()/dataset.shape[0])*100),1))
    carac['Proportion number of missing values(%)']=prop_missing_values
  #Moyenne
    moyenne=[]
    for i in dataset.columns.tolist():
        if is_numeric_dtype(dataset[i])==True:
            moyenne.append(dataset[i].mean())
        else:
            moyenne.append (np.nan)
    carac['moyenne']=moyenne
      #Maximum
    maxim=[]
    for i in dataset.columns.tolist():
        if is_numeric_dtype(dataset[i])==True:
            maxim.append(dataset[i].max())
        else:
            maxim.append (np.nan)
    carac['maximum']=maxim
      #Minimum
    minim=[]
    for i in dataset.columns.tolist():
        if is_numeric_dtype(dataset[i])==True:
            minim.append(dataset[i].min())
        else:
            minim.append (np.nan)
    carac['minimum']=minim
#Mode
    mode=[]
    for i in dataset.columns.tolist():
        mode.append(statistics.mode(dataset[i].tolist()))
    carac['mode']=mode
    prop_mode=[]
    for i in dataset.columns.tolist():
        if pd.isna(carac.loc[i,"mode"])==False:
            prop_mode.append(round((dataset[dataset[i]==carac.loc[i,"mode"]][i].value_counts().tolist()[0]/dataset[i].value_counts().sum())*100,1))
        else:
            prop_mode.append(round((dataset[i].isna().sum()/(dataset[i].value_counts().sum()+dataset[i].isna().sum()))*100,1))
    carac['proportion mode (%)']=prop_mode

#Afficher les caractéristiques
    return carac
```

```python
def milk_timesseries(dataset, date_var,nombres_lags):
    def lag_columns_by_place(dataset,nombres_lags):
        """
        create lags for all numerical variables in dataset
        """

        for i in dataset.select_dtypes(include= ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']).columns:
            if ("trend" not in i) and (i not in ['timestep','weekdays', 'day', 'month','day_of_month', 'start_of_month', 'days_in_month', 'end_of_month', 'day_of_year','week_of_year']):
                nom_col='previous'+'_'+i
                for j in range (1,nombres_lags+1):
                    dataset[nom_col +'_'+str(j)]=dataset.groupby('piezo_station_coordinates')[i].shift(j).fillna(method='bfill').sort_index()
        return(dataset)
    """
    Use the variable containing the date to create features
    """
        #Day of week
    dataset['day']= dataset[date_var].apply(lambda x:x.weekday()+1)
    # Day of the month
    dataset['day_of_month']=dataset[date_var].apply(lambda x :pd.to_numeric(x.strftime("%d")))
    #start of the month
    dataset['start_of_month']=[1 if dataset['day_of_month'][i]==1 else 0 for i in range(0,len(dataset[date_var]))]
    #days_in_month
    dataset['days_in_month']=[monthrange(dataset[date_var][i].year,dataset[date_var][i].month)[1] for i in range(0,len(dataset[date_var]))]
    #end of the month
    dataset['end_of_month']=[1 if dataset['day_of_month'][i]==dataset['days_in_month'][i]==1 else 0 for i in range(0,len(dataset[date_var]))]
    #day of year
    dataset['day_of_year']= dataset[date_var].apply(lambda x:x.timetuple().tm_yday)
    #week of year
    dataset['week_of_year']=dataset[date_var].apply(lambda x:x.isocalendar())
    #aadd lags
    dataset=lag_columns_by_place(dataset, date_var,nombres_lags)
    return(dataset)
```
